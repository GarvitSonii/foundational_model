```python
import torch
from pathlib import Path
import matplotlib.pyplot as plt

models_dir = Path("models")

epochs = []
ious = []

# iterate over all model_snn_*.pt files
for file in models_dir.glob("model_snn_*.pt"):
    ckpt = torch.load(file, map_location="cpu")
    epochs.append(ckpt['epoch'])
    ious.append(ckpt['iou'])

# sort by epoch
epochs, ious = zip(*sorted(zip(epochs, ious)))

# plot
plt.figure(figsize=(7,4))
plt.plot(epochs, ious, marker='o')
plt.xlabel("Epoch")
plt.ylabel("Validation IoU")
plt.title("IoU vs Epoch")
plt.grid(True)
plt.show()
```
