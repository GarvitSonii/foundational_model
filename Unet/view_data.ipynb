```python
import torch
import matplotlib.pyplot as plt
import numpy as np
```

```python
sample = torch.load("cached_dataset/train/000850.pt")   # any file
```

```python
img = sample['img'].numpy()   # shape (4,H,W)
mask = sample['mask'].numpy() # shape (1,H,W)
```

```python
R, G, B, N = img[0], img[1], img[2], img[3]
```

```python
rgb = np.stack([R, G, B], axis=-1)
```

```python
plt.figure(figsize=(5,5))

plt.subplot(1,2,1)
plt.imshow(rgb)
plt.title("RGB")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(mask[0], cmap='gray')
plt.title("Mask")
plt.axis("off")

plt.tight_layout()
plt.show()
```
